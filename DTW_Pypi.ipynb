```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from dtaidistance import dtw

sin_df = pd.read_csv('/content/sample_data/corr_data.csv')
print(sin_df.shape)

sin1 = np.array(sin_df['Sin wave 1'][:522]) # [:522]
sin2 = np.array(sin_df['Sin wave 2'][:522])

DTW_diff_Sin = dtw.distance_fast(sin1, sin2, use_pruning=True) #use_pruning=True window = 30
print('DTW_diff_Sin = ', DTW_diff_Sin)

plt.plot(np.arange(len(sin1)), sin1)
plt.plot(np.arange(len(sin2)), sin2)
plt.xlabel("Number of Data Points")
plt.ylabel("Sin Value")
plt.legend(["Sin Wave1", "Sin Wave2"])
plt.show()
```

```python
from dtaidistance import ed
def find_euclideanDistance_strokes(file1path, file2path):
  stroke_df_1 = pd.read_csv(file1path)
  stroke_df_2 = pd.read_csv(file2path)

  print(stroke_df_1.shape, stroke_df_2.shape)
  val = len(stroke_df_2) if len(stroke_df_2) < len(stroke_df_1) else len(stroke_df_1)
  stroke_Data_1 = np.array(stroke_df_1['Data'][:val])
  stroke_Data_2 = np.array(stroke_df_2['Data'][:val])

  sensor1, strok1, sensor2, strok2 = split_filePath(file1path, file2path)
  varName = "ED_diff_" + strok1 + "_" + strok2
  ed_distance = ed.distance(stroke_Data_1, stroke_Data_2)
  print(f'{varName} = ', ed_distance)

  
  plt.plot(np.arange(len(stroke_Data_1)), stroke_Data_1)
  plt.plot(np.arange(len(stroke_Data_2)), stroke_Data_2)
  plt.xlabel("Number of Data Points")
  plt.ylabel(sensor1 + " Value")
  plt.legend([sensor1+" "+strok1+" data", sensor2+" "+strok2+" data"])
  plt.show()

def split_filePath(file1path, file2path):
  file1Info = file1path.split('/')[-1].split('_')
  file2Info = file2path.split('/')[-1].split('_')
  return file1Info[0], file1Info[1], file2Info[0], file2Info[1]
```
